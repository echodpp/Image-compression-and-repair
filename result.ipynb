```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage, misc, signal
import mini_project as mp
```

## Fishing boat image
Using $8 \times 8$ block and sampling each block with range of 4 to 64 (out of 64)

```python
boat = mp.imgRead('fishing_boat.bmp', 'L')
sample_lst = [8 * i for i in range(1, 9)]
boat_lst = []
boat_filt_lst = []
```

```python
for s in sample_lst:
    print("Sampled size: {}".format(s))
    result = mp.imgRecover(boat, 8, s)
    imgFilt = mp.applyMedFilter(result)
    boat_lst.append(result)
    boat_filt_lst.append(imgFilt)
    mp.compareImg(boat, result, imgFilt, show_image=False)
    print('-------------------------------------------------------')
```

```python
n1 = 0
print('Sampled size: {} out of 64'.format(sample_lst[n1]))
mp.compareImg(boat, boat_lst[n1], boat_filt_lst[n1])
```

```python
for n1 in range(8):
    print('Sampled size: {} out of 64'.format(sample_lst[n1]))
    mp.compareImg(boat, boat_lst[n1], boat_filt_lst[n1])
```

### Block size 4x4

```python
sample_lst_4 = [2 * i for i in range(4,9)]
boat_lst_4 = []
boat_filt_lst_4= []
```

```python
for s in sample_lst_4:
    print("Sampled size: {}".format(s))
    result = mp.imgRecover(boat, 4, s)
    imgFilt = mp.applyMedFilter(result)
    boat_lst_4.append(result)
    boat_filt_lst_4.append(imgFilt)
    mp.compareImg(boat, result, imgFilt, show_image=False)
    print('-------------------------------------------------------')
```

```python


plt.figure(figsize=(18, 8))
for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.imshow(boat_lst_4[i], cmap='gray', vmin=0, vmax=255)
    plt.title("Restored Image")
    plt.subplot(2, 5, i+6)
    plt.imshow(boat_filt_lst_4[i], cmap='gray', vmin=0, vmax=255)
    plt.title("Restored Image + MedFilt")
```
